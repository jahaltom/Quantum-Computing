```python
import requests
import re

url = "https://beacon.nist.gov/beacon/2.0/pulse/last"  # NIST mirror for IDQ format
try:
    resp = requests.get(url)
    pulse = resp.json()
    output_value = pulse['pulse']['outputValue']
    timeStamp = pulse['pulse']['timeStamp']
    print("‚úÖ IDQ Entropy:", output_value)
    print("Time Stamp:",timeStamp )
except Exception as e:
    print("‚ö†Ô∏è Failed to fetch IDQ entropy:", e)
```

```python
import numpy as np

seed = int(output_value, 16)
rng = np.random.default_rng(seed)
print("üé≤ Quantum-seeded int:", rng.integers(0, 1000000))
```
